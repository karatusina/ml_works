```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn import model_selection
from sklearn import tree
import graphviz

from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.tree import DecisionTreeClassifier
import graphviz
```

```python
iris = datasets.load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['flower_species'] = df['target'].apply(lambda x: iris.target_names[x])
```

```python
df.head(2)
```

```python
X = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
y = df[['target']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# напишем разные критерии для дальнейшего перебора
params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [1, 2, 3, 4, 5],
    'min_samples_leaf': [1, 2, 3, 4, 5]
}
```

```python
model = DecisionTreeClassifier(random_state=1)
```

```python
grid_ = GridSearchCV(
    estimator=model,
    param_grid=params,
    scoring='accuracy',
    cv=5,
    verbose=1,
    n_jobs=-1
)
grid_.fit(X_train, y_train.values.ravel())
```

```python
model_b = grid_.best_estimator_
params_b = grid_.best_params_

test_score = model_b.score(X_test, y_test.values.ravel())

print("Лучшие параметры:", params_b)
print("Точность лучшей модели на тестовой выборке:", round(test_score, 2))
```
